### **1.1. Nativo** 

Neste primeira atividade, você deve ler o código abaixo implementado em Python nativo e responder a seguinte pergunta:

> O que o código abaixo computa?

**Resposta**: 
1. No primeiro bloco de código read, realiza a leitura do arquivo traffic.csv.
2. No segundo bloco de código analytics, passa para as variáveis:
       day = 14;
       incidents = 0;
       incident_by_day = dicionario;
realiza a interação em data separando por linhas puladas e depois passando a variável timebox_data = timebox, separando por ",".
3. Cria uma lista chamada incident e percorre a lista timebox_data, iniciando no índice 1 e percorrendo toda a lista pegando toda ela.
4. Cria list half_hour extrai o valor inteiro da posição 0,
realiza condição if para verificar se a hora é igual a 27, caso for day(data) vai ser igual a 14, insere na lista incidents
  - Após isso day(data) soma mais um.
  - incidents = 0
  except ValueError:
    continue
  - Em caso de erro de valor continua o código

  5. e última parte, realiza interação na lista incident_by_day, exibe os dados de quantidade de acidentes por dia.


```
# -- read

data = None

with open(file='traffic.csv', mode='r', encoding='utf8') as fp:
  
  fp.readline()
  data = fp.read()

# -- analytics

day = 14
incidents = 0
incident_by_day = dict()

for timebox in data.split(sep='\n'):

  timebox_data = timebox.split(sep=';')

  # --
  # -- inicio da computação escalar
  # --

  for incident in timebox_data[1: len(timebox_data)-1]:
    incidents = incidents + int(incident)

  # --
  # -- fim da computação escalar
  # --

  try: 

    half_hour = int(timebox_data[0])

    if half_hour == 27:
      incident_by_day[day] = incidents
      day = day + 1
      incidents = 0

  except ValueError:
    continue

# -- results

for day in incident_by_day:
  
  print(f'{day}: {incident_by_day[day]}')
```

### **1.2. NumPy** 

Substitua o trecho do código do algoritmo que utiliza da **computação escalar** por um que utiliza da **computação vetorial**. Use o pacote NumPy.

```
# -- read

data = None

with open(file='traffic.csv', mode='r', encoding='utf8') as fp:
  
  fp.readline()
  data = fp.read()

# -- analytics

import numpy as np


day = 14
incidents = 0
incident_by_day = dict()

for timebox in data.split(sep='\n'):

  timebox_data = timebox.split(sep=';')

  # --
  # -- inicio da computação vetorial
  # --
  
  incident = np.array(timebox_data)
  incidents = incidents + int(incident[1:-1].astype(float).sum())

  # --
  # -- fim da computação vetorial
  # --

  try: 

    half_hour = int(timebox_data[0])

    if half_hour == 27:
      incident_by_day[day] = incidents
      day = day + 1
      incidents = 0

  except ValueError:
    continue

# -- results

for day in incident_by_day:

  print(f'{day}: {incident_by_day[day]}')
```
