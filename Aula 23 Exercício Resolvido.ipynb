## 1\. Pinguins 

Neste exercício, vamos utilizar uma base de dados com informações sobre penguins. A idéia é preparar a base de dados para prever a espécie do penguin (variável resposta) baseado em suas características físicas e geográficas (variáveis preditivas).

```
import numpy as np
import pandas as pd
import seaborn as sns
```

```
data = sns.load_dataset('penguins')
```

```
data.head()
```

```
data.info()
```

```
data.describe()
```

### **1.1. Valores nulos** 

A base de dados possui valores faltantes, utilize os conceitos da aula para trata-los.

```
# resposta da questão 1.1

inserindo_media = data.fillna(data.mean())
inserindo_media
```

```
inserindo_media.info()
```

```
inserindo_media.describe()
```

```
removendo_colunas_ausentes = inserindo_media.dropna(axis=1)
```

```
removendo_colunas_ausentes
```

```
removendo_colunas_ausentes.info()
```

### **1.2. Variáveis numéricas** 

Identifique as variáveis numéricas e crie uma nova coluna **padronizando** seus valores. A nova coluna deve ter o mesmo nome da coluna original acrescidade de "*_std*".

> **Nota**: Você não deve tratar a variável resposta.

```
# resposta da questão 1.2

variáveis_numericas = removendo_colunas_ausentes


variáveis_numericas[['bill_length_mm','bill_depth_mm','flipper_length_mm','body_mass_g']].head()
```

```
min = variáveis_numericas['bill_length_mm'].min()
print(min)

max = variáveis_numericas['bill_length_mm'].max()
print(max)
```

```
media = variáveis_numericas['bill_length_mm'].mean()
print(media)

desvio_padrao = variáveis_numericas['bill_length_mm'].std()
print(desvio_padrao)
```

```
#Inserindo coluna bill_length_mm_std:
variáveis_numericas['bill_length_mm_std'] = variáveis_numericas['bill_length_mm'].apply(lambda calda: (calda - media) / desvio_padrao)
```

```
variáveis_numericas
```

```
min = variáveis_numericas['bill_depth_mm'].min()
print(min)

max = variáveis_numericas['bill_depth_mm'].max()
print(max)
```

```
media = variáveis_numericas['bill_depth_mm'].mean()
print(media)

desvio_padrao = variáveis_numericas['bill_depth_mm'].std()
print(desvio_padrao)
```

```
#Inserindo coluna bill_depth_mm_std:
variáveis_numericas['bill_depth_mm_std'] = variáveis_numericas['bill_depth_mm'].apply(lambda bico: (bico - media) / desvio_padrao)
variáveis_numericas
```

```
media = variáveis_numericas['flipper_length_mm'].mean()
print(media)

desvio_padrao = variáveis_numericas['flipper_length_mm'].std()
print(desvio_padrao)
```

```
#Inserindo coluna bill_depth_mm_std:
variáveis_numericas['flipper_length_mm_std'] = variáveis_numericas['flipper_length_mm'].apply(lambda nadadeira: (nadadeira - media) / desvio_padrao)
variáveis_numericas
```

```
media = variáveis_numericas['body_mass_g'].mean()
print(media)

desvio_padrao = variáveis_numericas['body_mass_g'].std()
print(desvio_padrao)
```

```
#Inserindo coluna bill_depth_mm_std:
variáveis_numericas['body_mass_g_std'] = variáveis_numericas['body_mass_g'].apply(lambda massa_corporal: (massa_corporal - media) / desvio_padrao)
variáveis_numericas
```

```
variáveis_numericas.info()
```

### **1.3. Variáveis categóricas** 

Identifique as variáveis categóricas nominais e ordinais, crie uma nova coluna aplicando a técnica correta de conversão a seus valores. A nova coluna deve ter o mesmo nome da coluna original acrescidade de "*_nom*" ou "*_ord*".

> **Nota**: Você não deve tratar a variável resposta.

```
# resposta da questão 1.3

variáveis_numericas['island'].drop_duplicates()
```

```
variáveis_numericas['island_Torgersen_nom'] = data['island'].apply(lambda ilha: 1 if ilha == 'Torgersen' else 0)
variáveis_numericas['island_Biscoe_nom'] = data['island'].apply(lambda ilha: 1 if ilha == 'Biscoe' else 0)
variáveis_numericas['island_Dream_nom'] = data['island'].apply(lambda ilha: 1 if ilha == 'Dream' else 0)
variáveis_numericas.head()
```

```
variáveis_numericas.info()
```

### **1.4. Limpeza** 

Descarte as colunas originais e mantenha apenas a variável resposta e as variáveis preditivas com o sufixo *_std*", *_nom*" e "*_ord*". 

```
# resposta da questão 1.4

data_final = variáveis_numericas[['species','bill_length_mm_std','bill_depth_mm_std',
                                  'flipper_length_mm_std','body_mass_g_std',
                                  'island_nom','island_Torgersen_nom',
                                  'island_Biscoe_nom','island_Dream_nom']]
data_final
```
